Let's read in the csv attached in the repository. 

It is the same data as in the response to the FOIA requests 53475 and 57818, with a few modifications:

- The response to 57818 specified that 3 children had been deported 2010 and 2011: this indication has been removed, but the deported children are counted among the total deportations. 

- The flight of 28/11/2019 was hypothesised to have been a deportation to the EU under Dublin Regulations (based on the destination and other flights in this period). 

- The flights of 22/05/12, 24/08/12, 14/12/12 to Italy, Austria, and Portugal were assumed to have been deportations under the Dublin Regulations. 

- 'D R Congo' was replaced by 'Congo (Democratic Republic)'. 

```python
# data: import data
data = pd.read_csv(
    './Deportation charter flights.csv',
    index_col=None,
    parse_dates=[0],
    dayfirst=True,
)
```

A first look at the data to ensure we're ready to start. 

```python
# data: have a first look at the data
data.head()
```

### Data cleaning

```python
# data: show basic information about the data
data.info()
```

We have a 408x8 table. 

The `Operation name` has some null values (which won't prove a problem), but otherwise there are no missing values. 

Let's take a closer look at the destination countries of the chartered flights. 

```python
# data: aggregate total removals by destination
data.groupby(
    by='Destination',
    as_index=False
).sum()[['Destination', 'Total Removal Directions issued']].sort_values(
    by='Total Removal Directions issued',
    ascending=False
).set_index('Destination')
```

Such a large number of unique destinations makes analysis difficult. We will group destinations into regions to make sense of the data. 

It is stated in the responses to the FOIA requests that destinations indicated with a `**` are deportations to EU countries, within the frame of the Dublin agreements: we can group these items in the region __EU (Dublin)__

We can see that flights have been sent to both `Albania` and to `Kosovo, Albania`: since we can't easily disaggregate flights towards Kosovo from those towards Albania, we will group these two destinations into the region __Albania, Kosovo__. 

We will also group the countries located in __Sub-Saharan Africa__ into a region. 

Some destinations with very low numbers of deportations do not belong to any of the above regions: we will group them into the region __Other__. 

The remaining destinations will be considered as regions unto themselves, since they account for a large number of deportations. 

```python
# data: group destinations into 'Region' categories
regions = {
    'Sub-Saharan Africa': [
        'Nigeria',
        'Ghana',
        'Cameroon',
        'Congo (Democratic Republic)',
        'Gambia',
        'Sierra Leone',
        'Guinea',
        'Togo',
        'Tanzania'
    ],
    'Albania, Kosovo': [
        'Albania',
        'Kosovo'
    ],
    'EU (Dublin)': [
        '**'
    ],
    'Other': [
        'Jordan',
        'Morocco',
        'Egypt',
        'Suriname'
    ]
}

data['Region'] = data['Destination']

for region in regions:
    for item in regions[region]:
        mask = data['Destination'].str.find(item) > -1
        data.loc[mask, 'Region'] = region
        
data.head()
```

We will shorten the columns' names to go a bit faster. We'll also add a column with the total of people removed, regardless of sex. 

```python
# data: rename columns
data = data.rename(
    columns={
        'Date of flight': 'Date',
        'Operation name': 'Flight_name',
        'Total Removal Directions issued': 'Deportation_directions',
        'Returnees withdrawn': 'Deportation_directions_withdrawn',
        'Returnees (Male)': 'Males',
        'Returnees (Female)': 'Females',
        'Criminality cases': 'FNOs'
    }
)

# data: create Total_deportations column
data['Total_deportations'] = data['Males'] + data['Females']

# data: rearrange columns
data = data.reindex(columns=[
    'Date',
    'Region',
    'Destination',
    'Flight_name',
    'Deportation_directions',
    'Deportation_directions_withdrawn',
    'Total_deportations',
    'Males',
    'Females',
    'FNOs'
])

data.head()
```

We know that our information covers the years 2010 to 2019. 

```python
# data: dates of first and last flights
print(
    'First flight:\t{}\nLast flight:\t{}'.format(
        data['Date'].min().strftime('%Y-%m-%d'),
        data['Date'].max().strftime('%Y-%m-%d')
    )
)
```

Let's set a distinct, destination-specific colour-coding, so that our charts can be more easily readable. 

```python
# data: create a unique palette for regions
palette = {}

palette_full = sns.color_palette(
    palette='colorblind', 
    n_colors=len(data['Region'].unique())
).as_hex()

for i, j in enumerate(sorted(data['Region'].unique())):
    palette[j] = palette_full[i]
    
palette
```

## Initial exploration

Let's gain a better understanding of our data. 

```python
# data: sum and count of total deportations
total_sum = data['Total_deportations'].sum()
total_count = data['Total_deportations'].count()

print('Total (sum): {}\nTotal (count): {}'.format(total_sum, total_count))
```

A total of __17,734__ people were deported through __408__ flights in 2010-2019. 

```python
# data: minimum, mean, and maximum values
data.describe().loc[[
    'min',
    'mean',
    'max'
]].round(0).astype(int).transpose()
```

```python
# data: extreme values
mins_and_maxes = data.loc[
    (data['Deportation_directions'] == 399)
    | (data['Deportation_directions_withdrawn'] == 313)
    | (data['Total_deportations'] == 86)
    | (data['Females'] == 17)
    | (data['FNOs'] == 49)
    | (data['Total_deportations'] == 1)
]

# data: some styling to better emphasise extreme values
mins_and_maxes.drop(
    columns=['Flight_name']
).style.apply(
    lambda x:
    [
        '''
        font-weight: bold;
        font-style: italic;
        color: red
        '''
        if i > 0 and (i == x.max() or i == x.min()) else '' for i in x
    ],
    axis=0,
    subset=data.describe().columns
).format({
    'Date': '{:%Y-%m-%d}'
}).hide_index()
```

In 2010-2019, the average number of people deported on any given charter flight was _43_, with a maximum of _86_, justifying, in a dramatic way, the use of 'mass deportation' as a synonym for charter flights by advocacy groups. 

The maximum number of deportation directions issued for a _single_ flight is *399*. For comparison, the most widespread aircraft in the world, the A320, sits 150 passengers. 

The average of deportation directions issued is *108*, and *64* deportation directions are withdrawn on average. 

Remarkably, on several occasions, a charter flight was employed for the deportation of a single person: disproportionate in its own way, this also echoes the criticism the Home Office received for the costs of its charter flights.

## Trends over time

### Preparing the data

We won't need the `Flight_name` column. 

```python
# data: create a dataframe for this chapter
data_time = data.drop(columns=['Flight_name'])

data_time.head()
```

### Yearly trends

Let's check whether there is a yearly evolution in 2010-2019. 

```python
# data: totals per year
yearly_totals = data_time.groupby(
    pd.DatetimeIndex(data_time['Date']).year
).sum().drop(columns=['Deportation_directions', 'Deportation_directions_withdrawn'])
yearly_totals.index.names = ['Years']
yearly_totals.transpose()
```

```python
# plot: deportations per year, faceted by population
columns = ['Total_deportations', 'Males', 'Females', 'FNOs']

chart = alt.Chart(data_time)\
.mark_line(point=True)\
.transform_fold(columns, as_=['Population', 'Value'])\
.encode(
    x=alt.X('year(Date):T', title=None),
    y=alt.Y('sum(Value):Q', title=None),
    color=alt.Color('Population:N', sort=columns, legend=None),
    strokeDash=alt.StrokeDash('Population:N', sort=columns, legend=None),
    tooltip='Population:N'
)\
.properties(width=300, height=150)\
.facet(
    facet=alt.Facet(
        'Population:N',
        title=None,
        sort=columns,
        header=alt.Header(labelAnchor='start', labelFontStyle='italic')
    ),
    columns=2,
    title={
        'text': [
            'Yearly deportations,',
            '2010-2019'
        ],
        'anchor': 'middle',
        'fontSize': 15,
        'dy': -10
    }
)\
.resolve_scale(y='independent')

chart
```

```python
# seaborn
# plot: deportations per year, faceted by population
fig = plt.figure(figsize=(10, 5))

for i, population in enumerate(yearly_totals.columns):
    ax = fig.add_subplot(2, 2, i + 1)
    yearly_totals[population].plot.line(
        style=['-', '--', '-.', ':'][i],
        ax=ax
    )

    ax.set_title(
        population,
        style='italic',
        loc='left'
    )

    if i < 2:
        ax.tick_params(labelbottom=False)
    ax.set_xlabel('')

fig.suptitle(
    'Yearly deportations,\n2010-2019',
    fontsize='x-large',
    weight='semibold',
    y=1.05
);
```

### Overal trends and central moving average

We will be doing some very basic time series analysis with the data in the following chapters, and the explanations offered here hold for the rest of the analysis. 

Moving averages are a time series analysis technique that calculates a series of averages of a window of values: by smoothing out short-term fluctuations and highlighting longer-term patterns, it allows to observe the trend of the time series. We will be using centred moving averages throughout the analysis, given that they have the benefits of displaying a small lag, being easy to calculate and understand, and closely approximating trends yielded by more complex methods. 

```python
# DATA
# data: monthly deportations
deportations_month = data_time.resample('MS', on='Date').sum().reset_index()
deportations_month = deportations_month.drop(
    columns=['Deportation_directions', 'Deportation_directions_withdrawn']
)


# PLOTS
columns = deportations_month.columns[1:].values

# plot: base plot
base = alt.Chart(deportations_month)\
.transform_fold(fold=columns, as_=['Population', 'Value'])

# # plot: monthly deportations
deportations_chart = base\
.mark_line(opacity=.6)\
.encode(
    x=alt.X('Date:T', title=None),
    y=alt.Y('Value:Q', title=None),
    color=alt.Color('Population:N', legend=None, sort=columns)
)

# plot: cma
deportations_cma_chart = base\
.mark_line()\
.transform_window(
    CMA='mean(Value)',
    frame=[-3, 3],
    groupby=['Population']
)\
.encode(
    x=alt.X('Date:T', title=None),
    y=alt.Y('CMA:Q', title=None),
    color=alt.value('red'),
    strokeWidth=alt.value(3),
    tooltip=(alt.Tooltip('Date:T', title='Month'), alt.Tooltip('CMA:Q', format='.0f'))
)

# plot: yearly total annotations
deportations_text_chart = base\
.mark_text(align='right', dx=27)\
.encode(
    x=alt.X('year(Date):T', title=None),
    y=alt.value(10),
    text=alt.Text('sum(Value):Q', format=''),
)

# plot: background for annotations
deportations_text_background_chart = base\
.mark_rect(cornerRadiusTopRight=5, cornerRadiusBottomRight=5)\
.encode(
    x='year(Date):T',
    x2='yearquarter(Date):T',
    y=alt.value(2),
    y2=alt.value(17),
    color=alt.value('lightgrey')
)

# plot: final chart
chart = alt.layer(
    deportations_chart,
    deportations_cma_chart,
    deportations_text_background_chart,
    deportations_text_chart
)\
.properties(width=400, height=200)\
.facet(
    facet=alt.Facet(
        'Population:N',
        title=None,
        sort=columns,
        header=alt.Header(labelAnchor='start', labelFontStyle='italic')
    ),
    columns=2,
    title={
        'text':[
            'Deportations by month, with yearly totals',
            'and centred moving average, 2010-2019'
        ],
        'anchor':'middle',
        'fontSize': 15,
        'dy': -10
    },
)\
.resolve_scale(y='independent')

chart
```

```python
# seaborn
# DATA
# data: monthly deportations
deportations_monthly = data_time.set_index('Date').resample('MS').sum()
deportations_monthly.drop(
    columns=['Deportation_directions', 'Deportation_directions_withdrawn'],
    inplace=True
)

# data: centred moving average
cma = deportations_monthly.rolling(
    window=7, center=True, min_periods=0
).mean()


# PLOTS
fig = plt.figure(figsize=(15, 10))

# plot: monthly deportations and cma
for i, j in enumerate(deportations_monthly):
    ax = fig.add_subplot(2, 2, i + 1)
    deportations_monthly[j].plot(
        color=sns.color_palette()[i],
        ax=ax
    )
    cma[j].plot(
        color='crimson',
        linewidth=3,
        ax=ax
    )
    for k in range(2010, 2020):
        ax.text(
            x=(k - 1970) * 12 + 6,
            y=ax.get_ylim()[1] * 0.9,
            s=yearly_totals.loc[k, j],
            horizontalalignment='center',
            bbox={
                'boxstyle': 'round',
                'edgecolor': 'skyblue',
                'facecolor': 'white'
            }
        )
    if i < 2:
        ax.tick_params(labelbottom=False)
    ax.set_xlabel('')
    ax.set_ylim(bottom=0)
    ax.set_xlim(right='2020-01-01')

    ax.set_title(j, weight='bold', style='italic', loc='left')

fig.suptitle(
    'Deportations by month, with yearly totals\nand centred moving average,\n2010-2019',
    fontsize='x-large',
    weight='semibold'
);
```

The trends observed here are the same as those evidenced in [Yearly trends](#Yearly-trends): peak in 2013, gradual drop between 2014 and mid-2015, then a stagnation, and a sudden final drop in 2019. 

Deportations of females grew between 2016-2018, but still remained well below their previous levels, before a drop in 2019. 

In contrast, deportations of foreign national offenders have been increasing, irregularly, throughout most of the 10-year period. 

We can also observe that the CMA suggests that the number of deportations is lower, or higher, than it is in reality: this is a compromise we need to make to be able to observe the data in a way that is easier to process and make sense of. 

Another benefit of the CMA is that it is not unbiased (from a point of view of opinion). Because deportations are generally not seen in a positive light, the use of the CMA might be criticised solely on the basis that it fails to accurately represent months with 0 deportations. A retort to this is that the CMA also fails to represent months with 300+ deportations, but that the purpose of the CMA is to facilitate analysis and visualisation, not quote point values.

### Correlation matrix

Let's have a look at our data's correlation matrix to check whether any substantial (>0.8) unexpected correlation exists between the various variables in the dataset. 

```python
# DATA
# data: pairwise correlation for data_directions columns
correlation = data_directions.corr(
    method='pearson', min_periods=100
).reset_index().melt(id_vars='index')

# data: dummy ordinal variable for sorting
sort = {
    'Deportation_directions': 0,
    'Deportation_directions_withdrawn': 1,
    'Total_deportations': 2,
    'Males': 3,
    'Females': 4,
    'FNOs': 5
}

# PLOTS
# plot: pairwise correlation heatmap
heatmap_chart = alt.Chart(correlation)\
.mark_rect(stroke='white')\
.transform_calculate(
    order_rows='%s [datum.index]' % sort,
    order_cols='%s [datum.variable]' % sort
)\
.transform_filter(alt.datum.order_rows <= alt.datum.order_cols)\
.encode(
    x=alt.X('index:N', title=None, sort=list(sort.keys())),
    y=alt.Y('variable:N', title=None, sort=list(sort.keys())),
    color=alt.Color('value:Q', legend=None, scale=alt.Scale(scheme='goldred')),
    tooltip=(
        alt.Tooltip('variable', title='Row'),
        alt.Tooltip('index', title='Column')
    )
)

# plot: pairwise correlation text
heatmap_text_chart = heatmap_chart\
.mark_text(fontSize=15)\
.encode(
    text=alt.Text('value:Q', format='.2f'),
    color=alt.value('white')
)

# plot: final chart
chart = alt.layer(
    heatmap_chart,
    heatmap_text_chart
)\
.properties(
    title={
        'text': 'Correlation matrix',
        'fontSize': 15,
        'dy': -10
    },
    width=300,
    height=300
)

chart
```

### Relation between deportation directions issued and those withdrawn

Let's approach the rate of withdrawal from another angle, by plotting the number of deportation directions withdrawn against the total of deportation directions issued. 

```python
# DATA
# data: regression equation and correlation coefficient
regression = np.polyfit(
    data_directions['Deportation_directions'],
    data_directions['Deportation_directions_withdrawn'],
    1
)
correlation_coefficient = data_directions['Deportation_directions'].corr(
    data_directions['Deportation_directions_withdrawn']
)


# PLOTS
# plot: scatter plot of deportation directions against deportations directions withdrawn
scatter_chart = alt.Chart(data_directions)\
.mark_circle()\
.encode(
    x=alt.X('Deportation_directions:Q'),
    y=alt.Y('Deportation_directions_withdrawn:Q')
)

# plot: regression line of deportation directions against deportations directions withdrawn
regression_chart = scatter_chart\
.transform_regression(
    'Deportation_directions',
    'Deportation_directions_withdrawn'
)\
.mark_line(color='darkred')

# plot: final chart
chart = alt.layer(
    scatter_chart,
    regression_chart
)\
.properties(
    title={
        'text': [
            'Relationship between deportation directions issued',
            'and those withdrawn, 2010-2019'
        ],
        'fontSize': 15,
        'dy': -10,
        'subtitle': [
            'y = {r[0]:.2f} * x {r[1]:=4.0f}'.format(r=regression),
            'Correlation coefficient: {:.2f}'.format(correlation_coefficient)
        ]
    },
    width=600
)

chart
```

```python
# seaborn
# PLOTS
fig, ax = plt.subplots(figsize=(10, 5))

# plot: linear regression plot
sns.regplot(
    x='Deportation_directions',
    y='Deportation_directions_withdrawn',
    data=data_directions,
    ci=None,
    scatter_kws={'edgecolor': 'white'},
    line_kws={'color': 'green'},
    ax=ax
)

# plot: regression equation and correlation coefficient
regression = np.polyfit(
    data_directions['Deportation_directions'],
    data_directions['Deportation_directions_withdrawn'],
    1
)
correlation_coefficient = data_directions['Deportation_directions_withdrawn'].corr(
    data_directions['Deportation_directions'])

ax.text(
    x=10,
    y=250,
    s='y = {r[0]:.2f} * x {r[1]:=4.0f}\nCorrelation coefficient: {c:.2f}'.format(
        r=regression,
        c=correlation_coefficient
    ),
    bbox={
        'boxstyle': 'round',
        'edgecolor': 'skyblue',
        'facecolor': 'white'
    }
)

ax.set_xlim(left=0, right=300)
ax.set_ylim(top=300)
ax.set_title(
    'Relationship between deportation directions issued\nand those withdrawn, 2010-2019',
    fontsize='x-large',
    weight='semibold'
);
```

Surprisingly, the regression line evidences a very clear linear relationship between the number of deportation directions issued and withdrawn: a close estimate of the number of withdrawn deportation directions for any given flight can be obtained by multiplying the number of deportation directions for that flight by _0.77_ and substracting _19_. The correlation between the two variables is almost perfect, at _0.96_. This is unexpected: one would have assumed that a large number of factors influence whether a deportation direction is withdrawn or not, and that these factors will vary a lot over time, but this is not what the data suggests. 

Over the years that the Home Office has been deporting people through charter flights, one would expect that the authority would have determined a way to streamline this process, if not to minimise the mental distress of detained migrants, at least to minimise the costs of an ineffective deportation process. 

And yet, this is not the case. A worrying interpretation of this is that the extent of the Home Office's concern for a practice widely likened by detainees and anti-detention charities to psychological torture is not sufficient to promote procedures that are less erroneous.

### Volatility of withdrawals

We can also observe that the amplitude of variations in the previous plot doesn't seem to be constant. We can estimate the volatility of the proportions of withdrawal with a rolling standard deviation, which measures the standard deviation of the proportions of withdrawals from their moving average value. 

```python
# DATA
# data: centred moving standard deviation of proportion of withdrawals
withdrawals_monthly['CM_STD'] = withdrawals_monthly['Proportion_withdrawn'].rolling(
    window=7, center=True, min_periods=0
).std(ddof=0)


# PLOTS
# plots: proportion of withdrawals of deportation directions
withdrawals_proportion_chart = alt.Chart(withdrawals_monthly)\
.mark_line(opacity=.8)\
.encode(
    x=alt.X('Date:T', title=None),
    y=alt.Y(
        'Proportion_withdrawn:Q',
        axis=alt.Axis(
            title='Proportion_withdrawn',
            titleColor='blue',
            format='%'
        )
    )
)

# plot: centred moving standard deviation of proportion of withdrawals
withdrawals_std_chart = alt.Chart(withdrawals_monthly)\
.mark_line(strokeDash=[8, 4], color='red')\
.encode(
    x=alt.X('Date:T', title=None),
    y=alt.Y(
        'CM_STD:Q',
        axis=alt.Axis(
            title='CM_STD',
            titleColor='red',
            format='%'
        )
    )
)

chart = alt.layer(
    withdrawals_proportion_chart,
    withdrawals_std_chart
)\
.resolve_scale(y='independent')\
.properties(
    title={
        'text': [
            'Monthly proportion of withdrawals of deportation directions',
            'with centred moving standard deviation, 2010-2019'
        ],
        'fontSize': 15,
        'dy': -10
    },
    width=600
)

chart
```

Higher volatility is not a positive, as it indicates that there is no improvement in the accuracy of procedures over time. This negative interpretation obviously presupposes that the Home Office has (i) an incentive to issue a number of deportation directions that will be close to the total of deportations (for the benefit of migrants' mental health and the costs involved in cancelling deportation directions) and (ii) that it has the ability to improve the quality of its operations (which is a fair assumption to have).

We can observe that there are clear differences between the rates of withdrawal of deportation directions between different destinations. Few deportation directions issued for flights to _Albania, Kosovo_ and _EU_ fail to be carried out: the median rates of withdrawal for these two destination are 37% and 35%, respectively. However, a majority of those issued for flights to _Sub-Saharan Africa_ (a median of 66%) or _Pakistan_ (65%) are consistently withdrawn. 

We can also observe evolutions in rates of withdrawals, which goes to further demonstrate the need to look at how data evolves over time. Rates of withdrawal for _Sub-Saharan Africa_ and _Jamaica_ have shifted upwards , while those for _Albania_ and _Kosovo_ have remained roughly the same. 

A number of factors influence whether a deportation direction will be carried out or withdrawn. 

Deportations require a bilateral agreement: the receiving country must consent to the deportation, the complex bureaucratic procedures of a deportation on one end of the flight and a reception on the other need to be actioned. Consequently, this chart might signify that deportation toward the EU is straight-forward, while Pakistan is less inclined or reactive to deportation requests. 

Deportation directions are also withdrawn following successful legal and procedural challenges (by the deportees, their solicitors, advocacy charities, etc.): while these factors might contribute to the evolutions above, it is unlikely that they would create such differentials (would charities that support Jamaican deportees be so much more effective than those that support Dublenees?). 

Another explanation, which is probably the most likely, is that the above is mostly explained by Home Office procedures: a Home Office that approaches deportation directions in a more aggressive way (issues more deportation directions, spends less time building their case, issues deportation directions to people who are unlikely to be deported, etc.) would see its rate of withdrawals go up. 

Further analysis will provide more insights and allow us to argue for one of these explanations.

### Map of deportations by destination

```python
# DATA
# data: calculate the centroids for world countries
countries = gpd.read_file('countries-50m.json')

countries['Lon'] = countries.centroid.x
countries['Lat'] = countries.centroid.y
countries.drop(columns=['id', 'geometry'], inplace=True)

# data: identify destination/region pairs in the data
destinations_regions = {}

for index, row in data_region.iterrows():
    destinations = row['Destination'].replace('**', '')
    for destination in destinations.split(', '):
        if not destinations_regions.get(destination):
            destinations_regions[destination] = row['Region']

destinations_regions = pd.DataFrame.from_dict(destinations_regions, orient='index').reset_index()
destinations_regions.columns = ['Destination', 'Region']

# data: create dataframe with countries, region, and coordinates for the data
destinations_regions = destinations_regions.merge(
    countries,
    how='inner',
    left_on='Destination',
    right_on='name'
).drop(columns=['name'])

destinations_regions.head()
```

```python
# PLOTS
# plot: background for map
sphere = alt.sphere()
sphere_chart = alt.Chart(sphere)\
.mark_geoshape(fill='lightgrey')

# plot: countries
countries = alt.topo_feature('countries-50m.json', 'countries')

# plot: all countries
countries_background_chart = alt.Chart(countries)\
.mark_geoshape(fill='white', stroke='lightgrey', opacity=.8)

countries_uk_chart = countries_background_chart\
.mark_geoshape(stroke='black', strokeWidth=3, opacity=1)\
.transform_filter('datum.properties.name == "United Kingdom"')\
.encode(tooltip=alt.value('United Kingdom'))

# plot: destinations
countries_chart = alt.Chart(countries)\
.mark_geoshape(stroke='black')\
.transform_lookup(
    lookup='properties.name',
    from_=alt.LookupData(
        data=destinations_regions,
        key='Destination',
        fields=['Region']
    )
)\
.transform_lookup(
    lookup='Region',
    from_=alt.LookupData(
        data=data_region.groupby('Region', as_index=False).sum(),
        key='Region',
        fields=['Total_deportations']
    )
)\
.transform_filter(alt.datum.Region)\
.encode(
    color=alt.Color('Region:N', legend=None),
    tooltip=[alt.Tooltip('properties.name:N', title='Country'), 'Region:N']
)

# plot: circles
countries_deportations_chart = alt.Chart(countries)\
.mark_circle()\
.transform_lookup(
    lookup='properties.name',
    from_=alt.LookupData(
        data=destinations_regions,
        key='Destination',
        fields=['Region', 'Lon', 'Lat']
    )
)\
.transform_lookup(
    lookup='Region',
    from_=alt.LookupData(
        data=data_region.groupby('Region', as_index=False).sum(),
        key='Region',
        fields=['Total_deportations']
    )
)\
.transform_filter(alt.datum.Region)\
.transform_aggregate(
    Longitude='median(Lon)',
    Latitude='median(Lat)',
    Deportations='sum(Total_deportations)',
    groupby=['Region']
)\
.encode(
    longitude='Longitude:Q',
    latitude='Latitude:Q',
    color=alt.Color('Region:N', scale=alt.Scale(range=['black']), legend=None),
    size=alt.Size('Deportations:Q', scale=alt.Scale(range=[-1, 3300]), legend=alt.Legend(orient='bottom')),
    tooltip=['Region:N', 'Deportations:Q']
)

countries_deportations_background_chart = countries_deportations_chart\
.transform_calculate(
    Longitude2='''datum.Longitude < -10 ? datum.Longitude + 20 : datum.Longitude - 20''',
    Latitude2='''
        datum.Latitude > 42 ? datum.Latitude + 6 : (
            datum.Longitude > 30 & datum.Latitude > 30 ? datum.Latitude + 7 : (
                datum.Latitude < 20 ? datum.Latitude + 8 : datum.Latitude - 8
            )
        )
    '''
)\
.mark_rect(cornerRadius=10, opacity=.5, stroke='white')\
.encode(
    longitude2='Longitude2:Q',
    latitude2='Latitude2:Q',
    color=alt.Color('Region:N', scale=alt.Scale(range=['grey']), legend=None),
    tooltip=[]
)

countries_deportations_text_chart = countries_deportations_chart\
.mark_text(align='left', baseline='top', dx=5, dy=5, lineBreak=';', fontStyle='bold')\
.transform_joinaggregate(Total='sum(Deportations)')\
.transform_calculate(
    Longitude_text='datum.Longitude < -10 ? datum.Longitude : datum.Longitude - 20',
    Latitude_text='''
        datum.Latitude > 42 ? datum.Latitude + 6 : (
            datum.Longitude > 30 & datum.Latitude > 30 ? datum.Latitude + 7 : (
                datum.Latitude < 20 ? datum.Latitude + 8 : datum.Latitude
            )
        )
    ''',
    Percentage='datum.Deportations / datum.Total',
    Label='datum.Region + ":;" + format(datum.Deportations, ",") + " deportations;" + format(datum.Percentage, ".0%") + " of total"'
)\
.encode(
    longitude='Longitude_text:Q',
    latitude='Latitude_text:Q',
    text=alt.Text('Label:N'),
    color=alt.value('white'),
    size=alt.value(10),
    tooltip=[]
)

# plot: final chart
chart = alt.layer(
    sphere_chart,
    countries_background_chart,
    countries_uk_chart,
    countries_chart,
    countries_deportations_background_chart,
    countries_deportations_chart,
    countries_deportations_text_chart,
)\
.project(
    'mercator',
    scale=300,
    center=[1.69, 32.42]
)\
.properties(
    title={
        'text': ['Total deportations by country of destination,', '2010-2019'],
        'anchor': 'start',
        'fontSize': 25,
        'dx': 25,
        'dy': 60
    },
    width=880,
    height=520
).resolve_scale(color='independent')

chart
```

```python
# folium
# DATA
# data: deportations by region (absolute and relative)
destinations_total = data_region.groupby(
    by='Region'
).sum()['Total_deportations']

destinations_proportion = destinations_total.div(
    destinations_total.sum(),
    axis=0
) * 100

coefficients = dict(
    zip(
        destinations_total.index,
        [0.5, 1.2, 1, 1, 1, 1, 1.5, 1, 1]
    )
)

# data: geojson with destinations' coordinates
with open('./maps.json') as geojson:
    maps = json.load(geojson)


# PLOTS
# plot: map chart of deportations per region
centre = (46.13417004624326, -0.41748046875)

# basemap
map = folium.Map(
    location=centre,
    tiles='CartoDB Positron No Labels',
    zoom_start=3,
    min_zoom=3,
    max_zoom=3,
    zoom_control=False,
    min_lat=centre[0],
    max_lat=centre[0],
    min_lon=centre[1],
    max_lon=centre[1],
    max_bounds=True
)

# the geocoded data that specifies the borders of relevant destinations
folium.GeoJson(
    maps,
    style_function=lambda feature: {
        'fillColor': palette[feature['properties']['name']],
        'color': palette[feature['properties']['name']],
        'weight': 2,
        'opacity': 0.2,
        'fillOpacity': 0.5
    },
    tooltip=folium.GeoJsonTooltip(['name'], labels=False)
).add_to(map)

# function to find the centre of a region area
def get_centroid(coordinates):
    lon, lat = [sum(i) / len(i) for i in zip(*coordinates[0])]
    return lon, lat

# annotation of the map with bars and text
for feature in maps['features']:
    region = feature['properties']['name']
    color = palette[region]

    geometry = feature['geometry']
    if geometry['type'] == 'Polygon':
        lon, lat = get_centroid(geometry['coordinates'])
    elif geometry['type'] == 'MultiPolygon':
        coordinates = []
        for i in geometry['coordinates']:
            coordinates.append(get_centroid(i))
        lon, lat = [sum(i) / len(i) for i in zip(*coordinates)]

    folium.Rectangle(
        bounds=[
            [lat - 1, lon],
            [lat+1, lon-destinations_proportion[region]]
        ],
        tooltip=region,
        color=color,
        weight=1,
        opacity=0.2,
        fill=True,
        fill_color=color,
        fill_opacity=0.8
    ).add_to(map)

    folium.Marker(
        location=[lat, lon-destinations_proportion[region]*coefficients[region]],
        icon=folium.DivIcon(
            html='''
            <div style="
                background-color: rgba(242, 242, 242, 0.5);
                border-radius: 5px 5px 5px 0px;
                box-shadow: 4px 4px 8px 0 rgba(0, 0, 0, 0.2), 6px 6px 20px 0 rgba(0, 0, 0, 0.19);
                border: 1px solid {};
                padding: 4px 5px;
                display: inline-block;
            ">
                <span>
                    <em><strong>{}</strong></em>:
                    <br /><strong>{}</strong> deportations
                    <br /><strong>{:.1f}%</strong> of total
                </span>
            </div>
            '''.format(
                color,
                region,
                destinations_total[region],
                destinations_proportion[region]
            ),
            icon_size=(130, 0),
            icon_anchor=(0, 70),

        )
    ).add_to(map)

folium.Marker(
    location=[70, -80],
    icon=folium.DivIcon(
        html='''
        <div style="
            background-color: rgba(242, 242, 242, 0.5);
            border-radius: 5px;
            box-shadow: 4px 4px 8px 0 rgba(0, 0, 0, 0.2), 6px 6px 20px 0 rgba(0, 0, 0, 0.19);
            border: 1px solid black;
            padding: 4px 5px;
            display: inline-block;
            font-size: 20px;
        ">
            <span>
                <strong>Total deportations by country of destination,</strong>
                <br /><strong>2010-2019</strong>
            </span>
        </div>
        ''',
        icon_size=(1000, 0),
        icon_anchor=(0, 0),
    )
).add_to(map)

display(map)
```

### Proportion of FNOs among the deported, per destination

```python
# DATA
# data: proportion of FNOs per flight per destination
fnos_proportions = pd.DataFrame({
    'Region': data_fnos['Region'],
    'Proportion of FNOs': data_fnos['FNOs'] / data_fnos['Total_deportations']
})

# data: proportion of FNOs deported per destination
fnos_proportion_destination = (np.divide(
    data_fnos.groupby('Region').sum()['FNOs'],
    data_fnos.groupby('Region').sum()['Total_deportations']
)).sort_values(ascending=False)

# data: overall proportion of FNOs
fnos_proportion_total = np.divide(
    data_fnos['FNOs'].sum(),
    data_fnos['Total_deportations'].sum()
)


# PLOTS
fig, ax = plt.subplots(figsize=(10, 5))
colors = [palette[i] for i in fnos_proportion_destination.index]

# plot: violin and jitter plots of the per-flight proportions of FNOs
sns.violinplot(
    x='Proportion of FNOs',
    y='Region',
    data=fnos_proportions,
    inner=None,
    color='.8',
    order=fnos_proportion_destination.index,
    ax=ax
)
sns.stripplot(
    x='Proportion of FNOs',
    y='Region',
    data=fnos_proportions,
    linewidth=1,
    order=fnos_proportion_destination.index,
    palette=colors,
    ax=ax
)

ax.text(
    x=max(fnos_proportions['Proportion of FNOs']) + 0.15,
    y=0 - 0.5,
    s='Per dest.:',
    fontdict={'style': 'italic', 'weight': 'semibold'},
    ha='left',
    va='center',
    size='small'
)
for i, j in enumerate(fnos_proportion_destination):
    ax.text(
        x=max(fnos_proportions['Proportion of FNOs']) + 0.15,
        y=i,
        s='{:.1%}'.format(j),
        ha='left',
        va='center',
        size='small'
    )

# plot: overall proportion of FNOs
ax.axvline(
    fnos_proportion_total,
    linestyle='-.',
    color='skyblue'
)
ax.text(
    x=fnos_proportion_total,
    y=len(fnos_proportion_destination) - 0.5,
    s='''
        Overall proportion of FNOs: {:.1%}
    '''.format(fnos_proportion_total),
    ha='left',
    va='center',
    size='small'
)

ax.xaxis.set_major_formatter(PercentFormatter(xmax=1))
ax.set_xlabel('')
ax.set_xlim(left=-0.10, right=1.10)
ax.set_ylabel('')
ax.margins(y=0.1)
ax.set_title(
    'Proportion of FNOs among the deported,\nby destination, 2010-2019',
    fontsize='x-large',
    weight='semibold',
    y=1.05
);
```

Three quarters of Jamaican nationals deported from the UK by charter flight were deported as offenders.

We can confirm that Amber Rudd had a very high rate of withdrawal of deportation directions. In view of the metrics for May and Javid, it seems unlikely that the geopolitical context during Rudd's term would have been unusually unconducive to fruitful immigration enforcement partnerships: a simpler interpretation is that Rudd was the most superficially aggressive, ie the least measured, Home Secretary. We saw earlier that the volatility in the withdrawals of deportation directions increased quite a lot over 2014-2018, especially between mid-2017 and 2018: this corresponds to Rudd's term of office. 

This chart also validates the observation that Rudd also targeted females more than either Javid or Patel. In fact, Rudd also targeted females more than May, if one only looks at May's term after 2014 (after the massive drop in deportations that occurred in 2013). 

Javid had the lowest rate of withdrawal of deportation directions among the last three Home Secretaries: with a success rate at less than 40%, however, his approach to immigration enforcement cannot be described as methodical and just. 

Finally, Javid also targeted FNOs much more than his precessors: 36% of deportees during his term were deported as offenders. It is important however not to interpret this rashly by absolving him of at least part of the trauma involved in charter flight deportations specifically, and in deportations in general. Indeed, the legislation that governs when offenders can be deported (called automatic deportation) gives so much leeway to the Home Secretary, and its use has so often been attacked as completely disproportionate, that this metric could very well stem from Javid's abuse of legislation.

```python
# bokeh
# DATA
# data: deportations of females per destination per month
females_region_month = data_hs.groupby('Region').resample(
    'MS', 
    on='Date'
).sum()['Females'].reset_index()

# data: deportations of females per destination per month, cma
females_region_month_cma = females_region_month.pivot(
    index='Date',
    columns='Region',
    values='Females'
).fillna(0).rolling(window=7, center=True, min_periods=0).mean()

# data: deportations of females per destination per month, relative
females_region_month_cma_proportion = females_region_month_cma.div(
    females_region_month_cma.sum(axis=1),
    axis=0
)


# PLOTS
colors = [palette[i] for i in females_region_month_cma.columns]

# plot: deportations of females per destination per month, cma
fig1 = figure(
    x_axis_type="datetime",
    height=400,
    width=600,
    tools=''
)

plot = fig1.varea_stack(
    stackers=females_region_month_cma.columns,
    x='Date',
    source=females_region_month_cma,
    color=colors,
    alpha=0.6
)

# plot: parameters for term of office lines
def term(home_secretary):
    term_of_office = Span(
        location=home_secretary['term_start'],
        dimension='height',
        line_color=home_secretary['color'],
        line_dash='dotted',
        line_width=3,
        line_alpha=1
    )
    return term_of_office

# plot: general elections and prime minister changes
for index, row in elections.iterrows():
    if row['election_type'] == 'local': 
        continue
    fig1.add_layout(
        Span(
            location=row['election_date'],
            dimension='height',
            line_color='white',
            line_width=3
        )
    )

# plot: photos and terms of office for each Home Secretary
for index, row in home_secretaries.iterrows():
    fig1.multi_polygons(
        xs=[[[[
            row['term_start'],
            row['term_start'],
            row['term_end'],
            row['term_end']
        ]]]],
        ys=[[[[20.5, 21.5, 21.5, 21]]]],
        fill_color=row['color'],
        line_color='white',
        line_width=2,
        alpha=0.8,
        level='underlay'
    )
    fig1.add_layout(term(row))
    home_secretary = ImageURL(
        url=[row['url']],
        x=row['term_start'] + relativedelta(weeks=1),
        y=21,
        anchor='top_left'
    )
    fig1.add_glyph(home_secretary)

fig1.x_range.start = datetime.strptime('2010-01-01', '%Y-%m-%d')
fig1.x_range.end = datetime.strptime('2020-12-01', '%Y-%m-%d')
fig1.xaxis.visible = False

fig1.y_range.start = 0

# plot: legend
legend = Legend(
    items=[
        (i, [j]) for i, j in zip(females_region_month_cma.columns, plot[1:])
    ],
    location='center',
    orientation='horizontal',
    glyph_width=18
)
fig1.add_layout(legend, 'below')

fig1.add_layout(
    Title(
        text='''
        Deportations of females per destination,\n2010-2019
        ''',
        text_font_size='15px'
    ),
    place='above'
)

# plot: yearly figures
for year in range(2010, 2020):
    yearly_total = Label(
        x=datetime.strptime(str(year) + '-06-16', '%Y-%m-%d'),
        y=21,
        text=str(
            data_hs.groupby(
                pd.DatetimeIndex(data_hs['Date']).year
            ).sum().loc[year, 'Females']
        ),
        text_font_size='14px',
        text_color='grey',
        text_font_style='bold',
        border_line_color='skyblue',
        background_fill_color='white'
    )
    fig1.add_layout(yearly_total)

# plot: deportations of females per destination per month, cma, proportion
fig2 = figure(
    x_axis_type="datetime",
    height=400,
    width=600,
    x_range=fig1.x_range,
    tools=''
)

fig2.varea_stack(
    stackers=females_region_month_cma_proportion.columns,
    x='Date',
    source=females_region_month_cma_proportion,
    color=colors,
    alpha=0.6
)

# plot: general elections and prime minister changes
for index, row in elections.iterrows():
    if row['election_type'] == 'local': 
        continue
    fig2.add_layout(
        Span(
            location=row['election_date'],
            dimension='height',
            line_color='white',
            line_width=3
        )
    )

for index, row in home_secretaries.iterrows():
    fig2.add_layout(term(row))

fig2.x_range.start = datetime.strptime('2010-01-01', '%Y-%m-%d')
fig2.y_range.start = 0
fig2.yaxis[0].formatter = NumeralTickFormatter(format='0%')

fig2.add_layout(
    Title(
        text='''
        Deportations of females per destination (proportion),\n2010-2019
        ''',
        text_font_size='15px'
    ),
    place='above'
)

show(column(fig1, fig2))
```

```python
# DATA
# data: deportation of FNOs by home secretary by region
fnos_home_sec_region = data_hs.groupby(
    ['Home Secretary', 'Region'], as_index=False
).sum()[['Home Secretary', 'Region', 'FNOs']]


# PLOTS
# plot: stacked bar chart of deportations of FNOs by home secretary by region
fnos_home_sec_region_chart = alt.Chart(fnos_home_sec_region)\
.mark_bar(stroke='white')\
.encode(
    x=alt.X('FNOs:Q', title=None, stack='normalize', axis=alt.Axis(format='%')),
    y=alt.Y('Home Secretary:N', title=None, sort=home_secretaries['name'].values),
    color=alt.Color('Region:N'),
    order='color_Region_sort_index:Q',
    tooltip=['Region:N', alt.Tooltip('FNOs:Q', format=',')]
)

# plot: annotations
fnos_home_sec_region_text_chart = fnos_home_sec_region_chart\
.mark_text(align='right', lineBreak=';', baseline='bottom', dx=-10, fontStyle='bold')\
.transform_joinaggregate(Total='sum(FNOs)', groupby=['Home Secretary'])\
.transform_calculate(Percentage='''
    datum.FNOs / datum.Total > .05
    ? (
        format(
            datum.FNOs / datum.Total, ".0%")
        ) + ";(" + format(
            datum.FNOs, ","
        ) + ")"
    : ""
''')\
.encode(
    text=alt.Text('Percentage:N'),
    color=alt.Color('Region:N', scale=alt.Scale(range=['white']), legend=None),
)

# plot: final chart
chart = alt.layer(
    fnos_home_sec_region_chart,
    fnos_home_sec_region_text_chart
)\
.resolve_scale(color='independent')\
.properties(
    title={
        'text': [
            'Prevalence of different regions in the deportation of FNOs',
            'across the entire term of each Home Secretary'
        ],
        'fontSize': 15,
        'dy': -10
    },
    width=600,
    height=200
)

chart
```

```python
# seaborn
# DATA
# data: deportation of FNOs by home secretary by region
destinations_fnos_per_home_secretary_per_year = pd.DataFrame(
    {
        row['name']: (
            data_hs[
                data_hs['Home Secretary'] == row['name']
            ].groupby('Region')['FNOs'].sum().astype(int)
        ) for index, row in home_secretaries.iterrows()
    }
).fillna(0).astype(int).transpose().iloc[::-1]

# data: deportation of FNOs by home secretary by region, proportion
destinations_fnos_proportion_per_home_secretary = pd.DataFrame(
    {
        row['name']: (
            data_hs[
                data_hs['Home Secretary'] == row['name']
            ].groupby('Region')['FNOs'].sum().astype(int).div(
                data_hs.groupby('Home Secretary').sum()['FNOs'][row['name']]
            ) * 100
        ) for index, row in home_secretaries.iterrows()
    }
).transpose().iloc[::-1]


# PLOTS
fig, ax = plt.subplots(figsize=(15, 5))
colors = [palette[i] for i in destinations_fnos_per_home_secretary_per_year.columns]

# plot: deportation of FNOs by home secretary by region
destinations_fnos_proportion_per_home_secretary.plot.barh(
    alpha=0.8,
    color=colors,
    stacked=True,
    ax=ax
)

ax.legend(
    loc='center',
    bbox_to_anchor=(0.5, -0.20),
    ncol=3,
    fancybox=True
)

for i, patch in enumerate(ax.patches):
    average = destinations_fnos_per_home_secretary_per_year.iloc[i % 4, i // 4]
    if patch.get_width() >= 3:
        ax.annotate(
            '{:.0f}%\n({})'.format(patch.get_width(), average),
            xy=(
                patch.get_x() + patch.get_width() / 2,
                patch.get_y() + patch.get_height() / 2),
            verticalalignment='center',
            horizontalalignment='center',
            color='white',
            fontsize='large'
        )

ax.set_title(
    'Prevalence of each destination\namong the deportations of FNOs\nacross the entire term of each Home Secretary',
    fontsize='x-large',
    weight='semibold'
);
```

```python
# DATA
# data: deportations of FNOs per destination per month
fnos_region_month = data_hs.groupby('Region').resample(
    'MS', on='Date'
).sum()['FNOs'].reset_index()

# data: deportations of FNOs per destination per month, cma
fnos_region_month_cma = fnos_region_month.pivot(
    index='Date',
    columns='Region',
    values='FNOs'
).fillna(0).rolling(
    window=7, center=True, min_periods=0
).mean().reset_index().melt('Date', value_name='FNOs')

# data: add column with home secretaries
for index, row in home_secretaries.iterrows():
    condition = (
        (fnos_region_month_cma['Date'] >= row['term_start']) 
        & (fnos_region_month_cma['Date'] <= row['term_end'])
    )
    fnos_region_month_cma.loc[condition, 'Home_Secretary'] = row['name']


# PLOTS
# plot: base chart
base = alt.Chart(fnos_region_month_cma)\
.mark_area(opacity=.8, stroke='white', strokeWidth=2)

# plot: deportations of FNOs per destination per month, cma
fnos_cma_chart = base\
.encode(
    x=alt.X('Date:T', title=None, axis=alt.Axis(labels=False, ticks=False)),
    y=alt.Y('FNOs:Q', title=None),
    color=alt.Color('Region:N', legend=None),
    tooltip=[
        'Region:N',
        alt.Tooltip('FNOs:Q', format='.0f'),
        'Home_Secretary:N'
    ]
)\
.properties(
    title={
        'text': [
            'FNOs per destination,',
            'centred moving average, 2010-2019'
        ],
        'fontSize': 15,
        'dy': -10
    },
    width=800,
    height=350
)

# plot: deportations of FNOs per destination per month, cma, stacked
fnos_cma_relative_chart = base\
.encode(
    x=alt.X('Date:T', title=None),
    y=alt.Y(
        'FNOs:Q',
        stack='normalize',
        title=None,
        axis=alt.Axis(format='%')
    ),
    color=alt.Color(
        'Region:N', legend=alt.Legend(
            orient='bottom', direction='vertical', columnPadding=40, columns=3
        )
    ),
    tooltip=[
        'Region:N',
        alt.Tooltip('FNOs:Q', format='.0f'),
        'Home_Secretary:N'
    ]
)\
.properties(
    title={
        'text': [
            'FNOs per destination (proportion),',
            'centred moving average, 2010-2019'
        ],
        'fontSize': 15,
        'dy': -5
    },
    width=800,
    height=350
)

# plot: home secretary terms
terms_chart = alt.Chart(home_secretaries)\
.mark_rule(size=4, strokeDash=[4, 1])\
.encode(
    x=alt.X('term_start:T'),
    color=alt.Color(
        'name:N',
        legend=None,
        scale=alt.Scale(
            domain=list(home_secretaries['name']),
            range=list(home_secretaries['color'])
        )
    )
)

# plot: general elections and prime minister changes
elections_chart = alt.Chart(elections)\
.mark_rule(size=2)\
.transform_filter('datum.election_type != "local"')\
.encode(
    x=alt.X('election_date:T'),
    color=alt.value('white'),
    tooltip='prime_minister:N'
)

# plot: portraits of home secretaries
faces_chart = alt.Chart(home_secretaries)\
.mark_image(height=100, width=100)\
.encode(
    x=alt.X('term_start:T'),
    y=alt.value(0),
    url=alt.Url('url:N'),
    tooltip='name:N'
)

# plot: final chart
chart = alt.vconcat(
    alt.layer(
        fnos_cma_chart,
        terms_chart,
        elections_chart,
        faces_chart
    ).resolve_scale(color='independent'),
    alt.layer(
        fnos_cma_relative_chart,
        terms_chart,
        elections_chart
    ).resolve_scale(color='independent')
)

chart
```
